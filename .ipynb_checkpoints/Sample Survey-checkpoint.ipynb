```python
import pandas as pd
import numpy as np
import random
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Generate sample data
data = {
    'ID': range(1, 501),
    'Age': np.random.randint(18, 80, 500),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], 500),
    'Ethnicity': np.random.choice(['White', 'Black', 'Asian', 'Hispanic', 'Other'], 500),
    'Income': np.random.randint(20000, 120000, 500),
    'Education Level': np.random.choice(['High School', 'Associate', 'Bachelor', 'Master', 'Doctorate'], 500),
    'Smoking Status': np.random.choice(['Never', 'Former', 'Current'], 500),
    'Physical Activity': np.random.randint(0, 7, 500),
    'BMI': np.random.uniform(18.5, 40.0, 500).round(1),
    'Chronic Conditions': np.random.choice(['None', 'Diabetes', 'Hypertension', 'Diabetes and Hypertension'], 500),
    'Mental Health Status': np.random.choice(['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'], 500),
    'Access to Healthcare': np.random.choice(['Easy', 'Moderate', 'Difficult'], 500),
    'Health Insurance': np.random.choice(['None', 'Private', 'Public'], 500),
    'Vaccination Status': np.random.choice(['Yes', 'No'], 500),
    'Health Satisfaction': np.random.randint(1, 11, 500)
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```

```python
# Save to CSV
df.to_csv('Wayne_County_Health_and_Wellbeing_Survey_2024.csv', index=False)

print("CSV file has been created.")
```
